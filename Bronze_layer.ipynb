```python
# CELL 1 — Imports & config
import pandas as pd
import os

# ⚠️ Change this to your actual folder path
BASE_PATH  = r"C:\Users\Lenovo\Downloads\nyc_taxi_project"
RAW_FILE   = os.path.join(BASE_PATH, "raw", "yellow_tripdata_2023-01.parquet")
BRONZE_DIR = os.path.join(BASE_PATH, "bronze")

os.makedirs(BRONZE_DIR, exist_ok=True)
print("Folders ready!")
```

```python
# CELL 2 — Read raw parquet file
df_raw = pd.read_parquet(RAW_FILE)

print(f"Total raw records : {len(df_raw):,}")
print(f"Total columns     : {len(df_raw.columns)}")
print(f"\nColumn names:")
for col in df_raw.columns:
    print(f"  - {col}")
```

```python
# CELL 3 — Preview data
print("First 5 rows:")
display(df_raw.head())

print("\nData types:")
print(df_raw.dtypes)

print("\nBasic statistics:")
display(df_raw.describe())
```

```python
# CELL 4 — Save as Bronze (raw, untouched)
BRONZE_PATH = os.path.join(BRONZE_DIR, "yellow_trips_bronze.parquet")
df_raw.to_parquet(BRONZE_PATH, index=False)

size_mb = os.path.getsize(BRONZE_PATH) / 1024 / 1024
print(f"Bronze saved!")
print(f"Path : {BRONZE_PATH}")
print(f"Size : {size_mb:.1f} MB")
```
